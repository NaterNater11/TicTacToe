```python
# Setting up environment
import pygame as py
import random
import time
```


# End Screen Code

```python
# End Screen Code
def End(condition):
    
    # Delay to Show End Board
    time.sleep(2)
    
    # Setting Screen's Text
    Exit_Game = font.render('Exit Game', True, 'black')
    Play_Again = font.render('Play Again', True, 'black')
    Statement = font.render(condition + ' has won', True, 'black')
    
    # Covers Game Screen
    screen.fill("beige")
    
    # Placing Buttons On The Screen
    down = height/4 *3 - 20
    py.draw.rect(screen, "white", [20, down, side, height/4], 0)
    py.draw.rect(screen, "black", [20, down, side, height/4], 2)
    py.draw.rect(screen, "white", [width-side-20 , down, side, height/4], 0)
    py.draw.rect(screen, "black", [width-side-20 , down, side, height/4], 2)
    
    # Creating Text For Screen
    screen.blit(Play_Again, (side/2, down+height/8))
    screen.blit(Exit_Game, (width-width/4, down+height/8))
    screen.blit(Statement, (width/3 + 20, height/2 - 20))

    # Updates Visuals
    py.display.update()
    
    # Loop For Player Input
    running = True
    while running:
        # Getting Mouse Coordinates
        a,b = py.mouse.get_pos()

        for event in py.event.get():
            # Shuts Down Pygame And Screen (1)
            if event.type == py.QUIT:
                py.quit()

            if event.type == py.MOUSEBUTTONDOWN:
                # Resets Board And Starts A New Game
                if (20 <= a <= side + 20) and (down <= b <= down + height/4):
                    Menu()
                
                # Program ends (1)
                if (width-side-20 <= a <= width-20) and (down <= b <= down + height/4):
                    py.quit()
```


# Shapes For Game Board

```python
# Function To Display Circle On Board
def circle(x, y):
    
    # Inner Circle Cuts Hole In a Circle
    py.draw.circle(screen, 'black', [x, y], int(width/12) , 0) 
    py.draw.circle(screen, 'beige', [x, y], int(height/12), 0) 
            
    py.display.update()
```

```python
# Function To Display Cross On Board
def cross(x, y):
    
    # Two Diagonal Lines
    py.draw.line(screen, 'black', [int(x-width/8), int(y-height/8)], [int(x+width/8), int(y+height/8)], int(height/15))
    py.draw.line(screen, 'black', [int(x-width/8), int(y+height/8)], [int(x+width/8), int(y-height/8)], int(height/15))
    
    py.display.update()
```

```python
# Observes State of Board to Determine Symbol Locations
def Display(board):
    place = 0
    
    # Loops Through All Board Loactions
    while place < 9:
        
        # Sets Y Coordinate With Appropriate Height
        if 0 <= place <= 2:
            y = height / 6
            
        if 3 <= place <= 5:
            y = height / 2
            
        if 6 <= place <= 8:
            y = 5* height/6
        
        # Sets X Coordinate With Appropriate Length
        if place == 0 or place == 3 or place == 6:
            x = width / 6
            
        if place == 1 or place == 4 or place == 7:
            x = width /2
            
        if place == 2 or place == 5 or place == 8:
            x = 5*width/6
        
        # X Found On Board and Calls Display Function
        if board[place] == 'X':
            cross(x, y)
            
        # O Found On Board and Calls Display Function
        if board[place] == 'O':
            circle(x, y)
            
        place += 1
```


# Checking Win Conditions

```python
# Checks Board to See if Game is Over
def Check(board):
    
    # Local Variable to Check
    win = False
    checks = ['X', 'O']
    place = 0
    
    while place < 2:
        
        # Checking Horizontal Win conditions
        row = 0
        while row < 9:
            
            # Brings Winning Symbol to End Screen
            if board[row] == checks[place] and board[row + 1] == checks[place] and board[row + 2] == checks[place]:
                End(checks[place])
            row += 3
            
            
        # Checking Vertical Win conditions
        column = 0
        while column < 3:
            
            # Brings Winning Symbol to End Screen
            if board[column] == checks[place] and board[column + 3] == checks[place] and board[column + 6] == checks[place]:
                End(checks[place])
            column += 1
            
        
        # Checking Diagonal Win conditions
        if board[0] == checks[place] and board[4] == checks[place] and board[8] == checks[place]:
                End(checks[place])
            
        elif board[2] == checks[place] and board[4] == checks[place] and board[6] == checks[place]:
                End(checks[place])
        
        place += 1
```


# Player Code

```python
# Player Selection
def Player(running, game, board, turn):
    # Player Plays as X
    spot = 9
    symbol = 'X'
    
    # Two Player Game, So Symbols Will Alternate
    if game == 1 and turn % 2 == 0:
        symbol = 'O'
    
    # Loop To Wait For Player's Selection
    while spot == 9 and running:
        a,b = py.mouse.get_pos()

        for event in py.event.get():
            # Instructions to end program (1)
            if event.type == py.QUIT:
                running = False
                py.quit()

            # Areas Corresponding With Board Placements, Change When Clicked (2)
            if event.type == py.MOUSEBUTTONDOWN:
                if (0 <= a <= width/3) and (0 <= b <= height/3):
                    spot = 0

                if (width/3 <= a <= 2*width/3) and (0 <= b <= height/3):
                    spot = 1

                if (2*width/3 <= a <= width) and (0 <= b <= height/3):
                    spot = 2

                if (0 <= a <= width/3) and (height/3 <= b <= 2*height/3):
                    spot = 3

                if (width/3 <= a <= 2*width/3) and (height/3 <= b <= 2*height/3):
                    spot = 4

                if (2*width/3 <= a <= width) and (height/3 <= b <= 2*height/3):
                    spot = 5

                if (0 <= a <= width/3) and (2*height/3 <= b <= height):
                    spot = 6

                if (width/3 <= a <= 2*width/3) and (2*height/3 <= b <= height):
                    spot = 7

                if (2*width/3 <= a <= width) and (2*height/3 <= b <= height):
                    spot = 8

        
    # Placement in Available and Updates Board 
    if board[spot] == 0:
        board[spot] = symbol
        return (board)

    # Spot has already been taken, Runs This Code Again
    else:
        Player(running, game, board, turn)
```


# Bot Playing

```python
# Bot picks random spot
def Random(board):
    # Generates Random number 0-8 (3)
    spot = random.randint(0, 8)

    # Checks Number's Corresponding Location
    if board[spot] == 0:
        # Updates Board
        board[spot] = 'O'
        return (board)
        
    # That Spot Has Already Been Taken
    else:
        Random(board)
```

```python
# Logic Bots
def Bot(board, game):
    
    # Hard Bot First Checks Their Locations
    if game == 4:
        symbol = 'O'
        
    # Bot Checks Opponents Locations
    if game == 3:
        symbol = 'X'
    
    # Checks for two of the same symbols in a row
    row = 0
    while row < 9:
        counter = 0
        
        if board[row] == symbol:
            counter += 1
            
        if board[row + 1] == symbol:
            counter += 1
            
        if board[row + 2] == symbol:
            counter += 1
            
        # Two Same Symbols Found, Places Remaining Location
        if counter >= 2:
            if board[row] == 0:
                board[row] = 'O'
                return board
                
            if board[row + 1] == 0:
                board[row + 1] = 'O'
                return board
            
            if board[row + 2] == 0:
                board[row + 2] = 'O'
                return board
            
        row += 3
    
    # Checks for two of the same symbols in a column
    column = 0
    while column < 3:
        counter = 0
        
        if board[column] == symbol:
            counter += 1
            
        if board[column + 3] == symbol:
            counter += 1
            
        if board[column + 6] == symbol:
            counter += 1
            
        # Two Same Symbols Found, Places Remaining Location
        if counter >= 2 and game == 3:
            if board[column] == 0:
                board[column] = 'O'
                return board
            
            if board[column + 3] == 0:
                board[column + 3] = 'O'
                return board
                
            if board[column + 6] == 0:
                board[column + 6] = 'O'
                return board
            
            return board
        
        column += 1
                
    # Checks Diagonals
    down = 0
    up = 0
    diagonal = [down, up]
    
    if board[0] == symbol:
        down += 1
        
    if board[8] == symbol:
        down += 1
        
    if board[4] == symbol:
        down += 1
        up += 1
        
    if board[2] == symbol:
        up += 1
        
    if board[6] == symbol:
        up += 1
        
    # Two Same Symbols Found, Places Remaining Location
    if up >= 2:
        if board[2] == 0:
            board[2] = 'O'
            return board

        if board[4] == 0:
            board[4] = 'O'
            return board

        if board[6] == 0:
            board[6] = 'O'
            return board
        
    # Two Same Symbols Found, Places Remaining Location
    if down >= 2:
        if board[0] == 0:
            board[0] = 'O'
            return board

        if board[4] == 0:
            board[4] = 'O'
            return board

        if board[8] == 0:
            board[8] = 'O'
            return board

    # No Winning Placements Found, Checks For Blocking Conditions
    if game == 4:
        Bot(board, 3)
        
    # No Blocking Positions Found, Random Picks Free Spot
    else:
        return (Random(board))
```


# Decides Who Plays

```python
# Central Station For Who Plays
def Board(running, turn, game, board):

    # Spaces left on board, For Tied Games
    spaces = 0
    while spaces < 9:
        # Load player's turn
        if game == 1 or turn % 2 != 0:
            board = Player(running, game, board, turn)

        # Easy Bot plays
        elif game == 2:
            board = Random(board)

        # Medium/Hard Bot plays
        else:
            Bot(board, game)

        # Updates Changed To the Board
        Display(board)
        # Checks to see if Game Continues
        Check(board)

        turn += 1
        spaces += 1

    # Game Has Tied
    End('Nobody')
```


# Flipping the Coin

```python
# Determines Who Will Play First
def Coin_Flip(running):
    
    # Randomly choses heads or tails
    flip = random.randint(1, 2)
    player = 0
    
    # Screen for coin flip
    screen.fill("beige")
    
    # Creates coin Visual
    py.draw.circle(screen, 'yellow', [width/2, height/3+15], (width + height)/7, 0) 
    py.draw.circle(screen, 'black', [width/2, height/3+15], (width + height)/7, 3) 
    
    # Button Display
    head = font.render('HEADS', True, 'black')
    tail = font.render('TAILS', True, 'black')
            
    # Adding Button Visuals
    down = height/4 *3 - 20
    py.draw.rect(screen, "white", [20, down, side, height/4], 0)
    py.draw.rect(screen, "black", [20, down, side, height/4], 2)
    py.draw.rect(screen, "white", [width-side-20 , down, side, height/4], 0)
    py.draw.rect(screen, "black", [width-side-20 , down, side, height/4], 2)
    # Adding Button Text
    screen.blit(head, (side/2, down+height/8))
    screen.blit(tail, (width-width/4, down+height/8))
    
    # Displaying Visuals
    py.display.update()

    # Waits For Player's Input
    while player == 0 and running:
        a,b = py.mouse.get_pos()

        # Closes Pygame and Screen (1)
        for event in py.event.get():
            if event.type == py.QUIT:
                py.quit()

            # Waits For Player to Pick Heads or Tails
            if event.type == py.MOUSEBUTTONDOWN:
                if (20 <= a <= side + 20) and (down <= b <= down + height/4):
                    player = 1

                if (width-side-20 <= a <= width-20) and (down <= b <= down + height/4):
                    player = 2



    # Coin Landed on Heads
    if flip == 1:
        screen.blit(head, (width/2-20, (width + height)/7-10))

    # Coin Landed on Tails
    else:
        screen.blit(tail, (width/2-20, (width + height)/7-10))


    # Tells User Whether They Won or Lost
    if flip == player:
        # Player Has Guessed the Flip Correctly
        value = ['first', 1]


    else:
        # Coin Landed on Other Side
        value = ['second', 2]

    text = font.render('You will play ' + value[0], True, 'black')

    #Display coin's value
    py.display.update()
    time.sleep(2)

    # Display player's turn
    screen.fill("beige")
    screen.blit(text, (width/3, height/2))
    py.display.update()
    time.sleep(2)

    return (value[1])
```


# Menu Screen

```python
# Basic Setup
py.init()
py.font.init()

# Screen Parameters
width = 600
height = 400
side = width/3 + 20

# Sets Up Window
screen = py.display.set_mode((width, height))
game1 = True

# Creating Font
font = py.font.SysFont('timesnewroman.ttf', int(width/20))

# Menu Screen
def Menu():
    
    # Instruction Variables
    running = True
    game = 0
    first = 1
    board = [0] * 9
    
    # Displaying Text
    text = font.render('Give me an X, Give me an O, Give me a Three in a Row', True, 'black')
    
    # Background
    screen.fill("beige")
    py.display.set_caption('Tic Tac Toe')
    screen.blit(text, (30, height-50))
    
    # Button Options
    n = 0
    game_modes = ["Normal (Two Player)", "Single Player (Easy)", "Single Player (Mid)", "Single Player (Hard)"]
    
    
    # Loop for efficiency
    # Horizontal placement
    change = width/2 - 20
    x = width/10
    while x <= width:
        
        # Vertical Placement
        y = height/10
        while y <= height:
            
            # Adding Button Visuals
            py.draw.rect(screen, "white", [x, y, side, height/4], 0)
            py.draw.rect(screen, "black", [x, y, side, height/4], 2)
            mode = font.render(game_modes[n], True, 'black')
            screen.blit(mode, (x+10, y + 30))
            
            n += 1
            y += height/2 
        x += change
    
    py.display.update()

    # Game Mode Selection
    while game == 0 and running:

        a,b = py.mouse.get_pos()
        # Instrcutions to End Program (Properly)
        for event in py.event.get():
            if event.type == py.QUIT:
                running = False

            # Player Selects Their Game Mode (2)
            if event.type == py.MOUSEBUTTONDOWN:
                if (width/10 <= a <= width/10 + side) and (height/10 <= b <= height/10 + height/4):
                    game = 1

                if (width/10 <= a <= width/10 + side) and (height/10 + height/2 <= b <= y - height/4):
                    game = 2

                if (x - change <= a <= x - change + side) and (height/10 <= b <= height/10 + height/4):
                    game = 3

                if (x - change <= a <= x - change + side) and (height/10 + height/2 <= b <= y - height/4):
                    game = 4

    # Determines If Bot Plays First
    if game != 1:
        first = Coin_Flip(running)

    # Print Background Colour and Grid for Game
    screen.fill("beige")

    # Vertical lines
    x = 0
    while x <= width:
        py.draw.line(screen, "black", (x, 0), (x, height))
        x += width/3 - 0.2

    # Horizontal lines
    y = 0
    while y <= height:
        py.draw.line(screen, "black", (0, y), (width, y))
        y += height/3 - 0.2

    py.display.update()

    # Visuals for screen
    while running:
        # Instrcutions to End Program
        for event in py.event.get():
            if event.type == py.QUIT:
                running = False
        Board(running, first, game, board)


                    
# Calls Menu Function Once
if game1 == True:
    game1 = False
    Menu()
    
    # Incase Game Doesn't Close
    py.quit()
```
