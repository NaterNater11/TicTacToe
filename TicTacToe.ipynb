```python
# Setting up environment
import pygame as py
import random
import time
```

```python
def Coin_Flip(running, font):
    
    # Randomly choses heads or tails
    flip = random.randint(1, 2)
    player = 0
    
    # Screen for coin flip
    screen.fill("beige")
    
    
    # Buttons
    

    
    
    # Simulates coin flip to determine who plays first

    while running and player == 0:
        # Instrcutions to end program
        for event in py.event.get():
            if event.type == py.QUIT:
                running = False

            # waits for player to pick heads or tails
      
    
    # Creates coin background
    py.draw.circle(screen, 'yellow', [300, 180], 150, 0) 
    py.draw.circle(screen, 'black', [300, 180], 150, 3) 
        
    # Coin landed on heads
    if flip == 1:
        head = font.render('HEADS', True, 'black')
        screen.blit(head, (300,180))
        
        
    # Coin landed on tails
    else:
        tail = font.render('TAILS', True, 'black')
        screen.blit(tail, (300,180))
        
    py.display.update()
        
        
    
    # Tells user whether they won or lost
    if flip == player:
        # Player has guessed the flip correctly
        text = font.render('You will play first', True, 'black')
        time.sleep(15)
        return (1)
        
    else:
        # Coin landed on other side
        text = font.render('You will play second', True, 'black')
        time.sleep(15)
        return (2)

```
