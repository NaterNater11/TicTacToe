```python
# Bot picks random spot
def Random(board):
    spot = random.randint(0, 8)
    
    # Checks for avaiblity
    if board[spot] == 0:
        board[spot] = 'O'
        
    # Finds Different Spot
    else:
        Random(board)
        
    return (board)
    
```

```python
def Bot(board, game):
    
    # Decides to win or block
    if game == 3:
        symbol = 'O'
        
    if game == 2:
        symbol = 'X'
        
    # Checks for two of the same symbols in a row
    row = 0
    while row < 9:
        counter = 0
        
        if board[row] == symbol:
            counter += 1
            
        if board[row + 1] == symbol:
            counter += 1
            
        if board[row + 2] == symbol:
            counter += 1
            
        # Bot Places winning piece
        if counter >= 2 and game == 3:
            board[row] = symbol
            board[row + 1] = symbol
            board[row + 2] = symbol
            
            return board
            
        # Bot Blocks player win
        elif counter >= 2 and game == 2:
            if board[row] == 0:
                board[row] = 'O'
                
            if board[row + 1] == 0:
                board[row + 1] = 'O'
                
            if board[row + 2] == 0:
                board[row + 2] = 'O'
                
            return board
        
        row += 3
    
    
    
    # Checks for two of the same symbols in a column
    column = 0
    while column < 3:
        counter = 0
        
        if board[column] == symbol:
            counter += 1
            
        if board[column + 3] == symbol:
            counter += 1
            
        if board[column] == symbol:
            counter += 1
            
        # Bot Places winning piece
        if counter >= 2 and game == 3:
            board[column] = symbol
            board[column + 3] = symbol
            board[column + 6] = symbol
            
            return board
            
        # Bot Blocks player win
        elif counter >= 2 and game == 2:
            if board[column] == 0:
                board[column] = 'O'
                
            if board[column + 3] == 0:
                board[column + 3] = 'O'
                
            if board[column + 6] == 0:
                board[column + 6] = 'O'
                
            return board
        
        column += 1
                
                
                
    # Checks Diagonals
    down = 0
    up = 0
    diagonal = [down, up]
    
    if board[0] == symbol:
        down += 1
        
    if board[8] == symbol:
        down += 1
        
    if board[4] == symbol:
        down += 1
        up += 1
        
    if board[2] == symbol:
        up += 1
        
    if board[6] == symbol:
        up += 1
        
        
    d = 0
    while d < 2
        # Bot Places winning piece
        if diagonal[d] >= 2 and game == 3:
            board[0] = symbol
            board[4] = symbol
            board[8] = symbol
            
            return board
            
        # Bot Blocks player win
        elif diagonal[d] >= 2 and game == 2:
            if board[2] == 0:
                board[2] = 'O'
                
            if board[4] == 0:
                board[4] = 'O'
                
            if board[6] == 0:
                board[6] = 'O'
                
            return board
        
        column += 1


    # No Winning Placements Found
    if game == 3:
        Bot(board, 2)
        
    # No Blocking Positions Found
    else:
        return (Random(Board))
```


# Decides Who Plays

```python
def Board(running, start, game, board):
    
    # Spaces left on board
    spaces = 9
    while spaces > 0:
         
        # Easy Bot plays
        if start % 2 == 0 and game == 2:
            board = Random(board)
            
        elif start % 2 == 0 and game != 1:
            Bot(board, game)
            
    
    
    # Last Call to check for win
    Check(board)
    End('Tied')
```


# Flipping the Coin

```python
def Coin_Flip(running):
    
    # Randomly choses heads or tails
    flip = random.randint(1, 2)
    player = 0
    
    # Screen for coin flip
    screen.fill("beige")
    
    # Creates coin background
    py.draw.circle(screen, 'yellow', [width/2, height/3+15], (width + height)/7, 0) 
    py.draw.circle(screen, 'black', [width/2, height/3+15], (width + height)/7, 3) 
    
    
    # Button Display
    head = font.render('HEADS', True, 'black')
    tail = font.render('TAILS', True, 'black')
            
    # Adding Visuals
    down = height/4 *3 - 20
    py.draw.rect(screen, "white", [20, down, side, height/4], 0)
    py.draw.rect(screen, "black", [20, down, side, height/4], 2)
    py.draw.rect(screen, "white", [width-side-20 , down, side, height/4], 0)
    py.draw.rect(screen, "black", [width-side-20 , down, side, height/4], 2)
    screen.blit(head, (side/2, down+height/8))
    screen.blit(tail, (width-width/4, down+height/8))


    
    py.display.update()
    
    # Simulates coin flip to determine who plays first

    while running:
        # Instrcutions to end program
        while player == 0 and running:
            a,b = py.mouse.get_pos()

            for event in py.event.get():
                if event.type == py.QUIT:
                    running = False

                # waits for player to pick heads or tails
                if event.type == py.MOUSEBUTTONDOWN:
                    if (20 <= a <= side + 20) and (down <= b <= down + height/4):
                        player = 1
                    
                    if (width-side-20 <= a <= width-20) and (down <= b <= down + height/4):
                        player = 2

        

        # Coin landed on heads
        if flip == 1:
            screen.blit(head, (width/2-20, (width + height)/7-10))


        # Coin landed on tails
        else:
            screen.blit(tail, (width/2-20, (width + height)/7-10))


        # Tells user whether they won or lost
        if flip == player:
            # Player has guessed the flip correctly
            value = 'first'
            

        else:
            # Coin landed on other side
            value = 'second'
        
        text = font.render('You will play ' + value, True, 'black')

        # Closing tab doesn't wait for timers
        if running:
            #Display coin's value
            py.display.update()
            time.sleep(5)
            screen.fill("beige")

            # Display player's turn
            screen.blit(text, (width/3, height/2))
            py.display.update()
            time.sleep(5)
            
        return (value)
```


# Menu Screen

```python
# Basic Setup
py.init()
py.font.init()

# Screen Parameters
width = 600
height = 400
side = width/3 + 20

screen = py.display.set_mode((width, height))
game1 = True

# Creating font
font = py.font.SysFont('timesnewroman.ttf', int(width/20))


# Menu Screen
def Menu():
    
    # Instruction Variables
    running = True
    game = 0
    board = [0] * 9
    
    
    # Displaying Text
    text = font.render('Give me an X, Give me an O, Give me a Three in a Row', True, 'black')
    
    
    # Background
    screen.fill("beige")
    py.display.set_caption('Tic Tac Toe')
    screen.blit(text, (30, height-50))
    
    
    # Button Display
    
    n = 0
    game_modes = ["Normal (Two Player)", "Single Player (Easy)", "Single Player (Mid)", "Single Player (Hard)"]
    
    
    # Loop for efficiency
    # Horizontal placement
    change = width/2 - 20
    x = width/10
    while x <= width:
        
        # Vertical Placement
        y = height/10
        while y <= height:
            
            # Adding Visuals
            py.draw.rect(screen, "white", [x, y, side, height/4], 0)
            py.draw.rect(screen, "black", [x, y, side, height/4], 2)
            mode = font.render(game_modes[n], True, 'black')
            screen.blit(mode, (x+10, y + 30))
            
            n += 1
            y += height/2 
        x += change
    
    py.display.update()


    # Gamemode selection
    while running:
        while game == 0 and running:
            
            a,b = py.mouse.get_pos()
            # Instrcutions to end program
            for event in py.event.get():
                if event.type == py.QUIT:
                    running = False
                    
                # Game Selection
                # NOTE: I dislike this format but a loop may prevent selection
                if event.type == py.MOUSEBUTTONDOWN:
                    if (width/10 <= a <= width/10 + side) and (height/10 <= b <= height/10 + height/4):
                        game = 1

                    if (width/10 <= a <= width/10 + side) and (height/10 + height/2 <= b <= y - height/4):
                        game = 2

                    if (x - change <= a <= x - change + side) and (height/10 <= b <= height/10 + height/4):
                        game = 3

                    if (x - change <= a <= x - change + side) and (height/10 + height/2 <= b <= y - height/4):
                        game = 4
                
        #Calls coin flip function
        if game != 1:
            first = Coin_Flip(running)


        # Print background colour and grid for game
        screen.fill("beige")

        # Vertical lines
        x = 0
        while x <= width:
            py.draw.line(screen, "black", (x, 0), (x, height))
            x += width/3 - 0.2

        # Horizontal lines
        y = 0
        while y <= height:
            py.draw.line(screen, "black", (0, y), (width, y))
            y += height/3 - 0.2

        py.display.update()

        # Visuals for screen0
        while running:
            # Instrcutions to end program
            for event in py.event.get():
                if event.type == py.QUIT:
                    running = False
    Board(running, first, game, board)


                    
# ending game
if game1 == True:
    Menu()
    game1 = False

py.quit()
```
